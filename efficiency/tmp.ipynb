```python
def phi(m, main_func):
    # (\v5 -> (\v6 -> (\v7 -> ((v6 = v4) ? v7 : ((v5 $ (v6 + 1)) $ (((v3 $ v6) > (v6 - 1)) ? (((v4 % v6) = 0) ? ((v7 / (v3 $ v6)) * ((v3 $ v6) - 1)) : v7) : v7))))))
    result = m
    for i in range(2, m + 1):
        if main_func(i) > i - 1:
            if m % i == 0:
                result = (result // main_func(i)) * (main_func(i) - 1)
    return result

# メイン関数
def main(n):
    # ((\v1 -> (\v2 -> ((v1 < v2) ? v1 : v2))) $ v4) $ 1
    min_val = min(n, 1)
    
    if n > 2:
        return min_val + phi(n, main)
    else:
        return min_val + n

for i in range(10):
    print(main(i))

```
